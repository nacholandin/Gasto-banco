```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_excel("BBVA.xlsx")
```

```python
df['Fecha'] = pd.to_datetime(df['Fecha'], format='%d/%m/%Y')
df['año'] = df['Fecha'].dt.year
df['dia'] = df['Fecha'].dt.day_name()
df['mes'] = df['Fecha'].dt.month_name()
df.head()
```

```python
ingresos = df.loc[df['Importe'] > 0, 'Importe'].sum()
gastos = df.loc[df['Importe'] < 0, 'Importe'].sum()
gastos = abs(gastos)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.to_csv('bbva.csv', index=False)
```

```python
resumen = pd.DataFrame({
    'Tipo': ['Ingresos', 'Gastos'],
    'Importe': [ingresos, gastos]
})

fig1 = px.bar(
    resumen,
    x='Tipo',
    y='Importe',
    color="Tipo",
    title='Ingresos vs Gastos'
)

fig1.show()
```

```python
df['Importe'] = df['Importe'].abs()
```

```python
df_agrupado = (df.groupby("Gasto", as_index=False)["Importe"].sum())

fig2 = px.bar(
    df_agrupado,
    x="Gasto",
    y="Importe",
    color="Gasto",   # color distinto por categoría
    title="Gasto total por categoría"
)

fig2.show()
```

```python
df_hogar = (
    df[df["Gasto"] == "Hogar"]
    .groupby("Categoria", as_index=False)["Importe"]
    .sum()
)

fig3 = px.bar(
    df_hogar,
    x="Categoria",
    y="Importe",
    color="Categoria",
    title="Gasto Hogar por Categoría"
)
fig3.show()
```

```python
df_niñas = (
    df[df["Gasto"] == "Niñas"]
    .groupby("Categoria", as_index=False)["Importe"]
    .sum())
fig4 = px.bar(
    df_niñas,
    x="Categoria",
    y="Importe",
    color="Categoria",
    title="Gasto Niñas por Categoría")

fig4.show()
```

```python
df_coche = (
    df[df["Gasto"] == "Coche"]
    .groupby("Categoria", as_index=False)["Importe"]
    .sum())
fig5 = px.bar(
    df_coche,
    x="Categoria",
    y="Importe",
    color="Categoria",
    title="Gasto coche por Categoría")

fig5.show()
```

```python
df_varios = (
    df[df["Gasto"] == "Varios"]
    .groupby("Categoria", as_index=False)["Importe"]
    .sum())
fig6 = px.bar(
    df_varios,
    x="Categoria",
    y="Importe",
    color="Categoria",
    title="Gasto coche por Categoría")

fig6.show()
```
